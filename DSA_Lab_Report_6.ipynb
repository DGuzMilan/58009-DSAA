```python
class Node:
  def __init__(self, data = None, next = None):
    self.data = data
    self.next = next

def insertNode(number):
  global head, rear
  temp = Node(number)
  if head is None:
    head = rear = temp
  else:
    rear.next = temp
    rear = temp

def searchNode(key):
  global head
  current = head
  prev = None
  while current is not None:
    if current.data == key:
      if prev is not None:
        prev.next = current.next
        current.next = head
        head = current
      return True
    prev = current
    current = current.next
  return False

head = None
rear = None

def display():
  global head
  if head is None:
    print("List is empty")
    return
  temp = head
  print("List: ", end="")
  while temp is not None:
    print(temp.data, end="")
    if temp.next is not None:
      print(" -> ", end="")
    temp = temp.next
  print("\n")

if __name__ == "__main__":
  insertNode("a")
  insertNode("b")
  insertNode("c")
  insertNode("d")
  insertNode("e")
  insertNode("f")
  display()

  while True:
    letter = (input("\nEnter the letter to be searched: "))
    if searchNode(letter):
      print("Searched:", letter)
    else:
      print("Not found!")
      break
    display()
```
