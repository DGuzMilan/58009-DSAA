```python
class Node:
  def __init__(self, data = None, next = None):
    self.data = data
    self.next = next

def insertNode(number):
  global head, rear
  temp = Node(number)
  if head is None:
    head = rear = temp
  else:
    rear.next = temp
    rear = temp

def searchNode(key):
  global head
  current = head
  prev = None
  while current is not None:
    if current.data == key:
      if prev is not None:
        prev.next = current.next
        current.next = head
        head = current
      return True
    prev = current
    current = current.next
  return False

head = None
rear = None

def display():
  global head
  if head is None:
    print("List is empty")
    return
  temp = head
  print("List: ", end="")
  while temp is not None:
    print(temp.data, end="")
    if temp.next is not None:
      print(" -> ", end="")
    temp = temp.next
  print("\n")

if __name__ == "__main__":
  insertNode("a")
  insertNode("b")
  insertNode("c")
  insertNode("d")
  insertNode("e")
  insertNode("f")
  display()

  while True:
    letter = (input("\nEnter the letter to be searched: "))
    if searchNode(letter):
      print("Searched:", letter)
    else:
      print("Not found!")
      break
    display()
```

```python
class Node:
  def __init__(self, data = None, next = None, prev = None):
    self.data = data
    self.next = None
    self.prev = None

class DoublyLinkedList:
  def __init__(self):
    self.head = None
    self.rear = None
    self.count = 0

  def appendItem(self, data):
    new_node = Node(data)
    if not self.head:
      self.head = new_node
      self.rear = new_node
    else:
      new_node.prev = self.rear
      self.rear.next = new_node
      self.rear = new_node
    self.count += 1

  def insertItem(self, data):
    new_node = Node(data)
    if not self.head:
      self.head = new_node
      self.rear = new_node
    else:
      new_node.next = self.head
      self.head.prev = new_node
      self.head = new_node
    self.count += 1

  def reverse(self):
    current = self.head
    while current:
      temp = current.next
      current.next = current.prev
      current.prev = temp
      current = temp
    temp = self.head
    self.head = self.rear
    self.rear = temp

  def display(self):
    current = self.head
    while current:
      print(current.data, end = "  ")
      current = current.next
    print()

  def countItem(self):
    return self.count

if __name__ == "__main__":
    sequence = DoublyLinkedList()
    sequence.appendItem(1)
    sequence.appendItem(2)
    sequence.appendItem(3)
    sequence.appendItem(4)
    print("Doubly Linked List:")
    sequence.display()

    itemEnd = int(input("Enter item to append at the end: "))
    sequence.appendItem(itemEnd)
    print("Updated Doubly Linked List:")
    sequence.display()
    print("Number of items in the list:", sequence.countItem())

    itemStart = int(input("Enter item to insert at the front: "))
    sequence.insertItem(itemStart)
    print("Updated Doubly Linked List:")
    sequence.display()

    sequence.reverse()
    print("Reversed Doubly Linked List:")
    sequence.display()
    print("Number of items in the list:", sequence.countItem())
```
